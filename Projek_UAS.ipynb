```python
import json
from abc import ABC, abstractmethod

# --- KELAS-KELAS BARANG ELEKTRONIK (sama seperti sebelumnya, untuk kelengkapan kode) ---

# Level 1: Kelas Abstrak Paling Dasar
class BarangElektronik(ABC):
    def __init__(self, nama: str, harga: float):
        self.__nama = nama
        self.__harga = harga
        self.__is_on = False

    @property
    def nama(self) -> str:
        return self.__nama

    @property
    def harga(self) -> float:
        return self.__harga

    @property
    def is_on(self) -> bool:
        return self.__is_on

    def _set_is_on(self, status: bool):
        self.__is_on = status

    @abstractmethod
    def hidupkan(self):
        pass

    @abstractmethod
    def matikan(self):
        pass

    def tampilkan_info(self):
        print(f"Nama: {self.nama}")
        print(f"Harga: Rp {self.harga:,.2f}")
        print(f"Status: {'Menyala' if self.is_on else 'Mati'}")

# Level 2: Kelas Abstrak Spesifik
class PerangkatRumahTangga(BarangElektronik):
    def __init__(self, nama: str, harga: float, daya_watt: int):
        super().__init__(nama, harga)
        self.__daya_watt = daya_watt

    @property
    def daya_watt(self) -> int:
        return self.__daya_watt

    @daya_watt.setter
    def daya_watt(self, watt: int):
        if watt > 0:
            self.__daya_watt = watt
        else:
            print("Daya listrik tidak bisa negatif.")

    def tampilkan_info(self):
        super().tampilkan_info()
        print(f"Daya Listrik: {self.daya_watt} Watt")

class PerangkatGenggam(BarangElektronik):
    def __init__(self, nama: str, harga: float, kapasitas_baterai_mah: int):
        super().__init__(nama, harga)
        self.__kapasitas_baterai_mah = kapasitas_baterai_mah

    @property
    def kapasitas_baterai_mah(self) -> int:
        return self.__kapasitas_baterai_mah

    @kapasitas_baterai_mah.setter
    def kapasitas_baterai_mah(self, mah: int):
        if mah > 0:
            self.__kapasitas_baterai_mah = mah
        else:
            print("Kapasitas baterai harus positif.")

    def tampilkan_info(self):
        super().tampilkan_info()
        print(f"Kapasitas Baterai: {self.kapasitas_baterai_mah} mAh")

# Level 3: Kelas Konkret
class Televisi(PerangkatRumahTangga):
    def __init__(self, nama: str, harga: float, daya_watt: int, ukuran_layar_inci: int):
        super().__init__(nama, harga, daya_watt)
        self.__ukuran_layar_inci = ukuran_layar_inci

    @property
    def ukuran_layar_inci(self) -> int:
        return self.__ukuran_layar_inci

    @ukuran_layar_inci.setter
    def ukuran_layar_inci(self, ukuran: int):
        if 10 <= ukuran <= 100:
            self.__ukuran_layar_inci = ukuran
        else:
            print("Ukuran layar TV tidak valid (10-100 inci).")

    def hidupkan(self):
        if not self.is_on:
            print(f"{self.nama} sedang dihidupkan. Selamat menonton!")
            self._set_is_on(True)
        else:
            print(f"{self.nama} sudah menyala.")

    def matikan(self):
        if self.is_on:
            print(f"{self.nama} sedang dimatikan. Sampai jumpa!")
            self._set_is_on(False)
        else:
            print(f"{self.nama} sudah mati.")

    def ganti_channel(self, channel: str):
        if self.is_on:
            print(f"{self.nama} mengganti channel ke: {channel}")
        else:
            print(f"{self.nama} mati. Hidupkan dulu untuk ganti channel.")

    def tampilkan_info(self):
        super().tampilkan_info()
        print(f"Ukuran Layar: {self.ukuran_layar_inci} inci")


class Kulkas(PerangkatRumahTangga):
    def __init__(self, nama: str, harga: float, daya_watt: int, kapasitas_liter: int):
        super().__init__(nama, harga, daya_watt)
        self.__kapasitas_liter = kapasitas_liter # Initialize the attribute
        self._set_is_on(True)

    @property
    def kapasitas_liter(self) -> int:
        return self.__kapasitas_liter

    @kapasitas_liter.setter
    def kapasitas_liter(self, liter: int):
        if liter > 0:
            self.__kapasitas_liter = liter
        else:
            print("Kapasitas kulkas harus positif.")

    def hidupkan(self):
        if not self.is_on:
            print(f"{self.nama} mulai mendinginkan. Siap menyimpan makanan!")
            self._set_is_on(True)
        else:
            print(f"{self.nama} sudah menyala dan mendinginkan.")

    def matikan(self):
        if self.is_on:
            print(f"{self.nama} berhenti mendinginkan. Pastikan makanan aman!")
            self._set_is_on(False)
        else:
            print(f"{self.nama} sudah mati.")

    def atur_suhu(self, suhu: int):
        if self.is_on:
            print(f"{self.nama} mengatur suhu ke {suhu}Â°C.")
        else:
            print(f"{self.nama} mati. Hidupkan dulu untuk mengatur suhu.")

    def tampilkan_info(self):
        super().tampilkan_info()
        print(f"Kapasitas: {self.kapasitas_liter} liter")


class Smartphone(PerangkatGenggam):
    def __init__(self, nama: str, harga: float, kapasitas_baterai_mah: int, sistem_operasi: str):
        super().__init__(nama, harga, kapasitas_baterai_mah)
        self.__sistem_operasi = sistem_operasi

    @property
    def sistem_operasi(self) -> str:
        return self.__sistem_operasi

    @sistem_operasi.setter
    def sistem_operasi(self, os: str):
        if os:
            self.__sistem_operasi = os
        else:
            print("Sistem operasi tidak boleh kosong.")

    def hidupkan(self):
        if not self.is_on:
            print(f"{self.nama} sedang booting. Selamat datang!")
            self._set_is_on(True)
        else:
            print(f"{self.nama} sudah menyala.")

    def matikan(self):
        if self.is_on:
            print(f"{self.nama} sedang dimatikan. Sampai jumpa!")
            self._set_is_on(False)
        else:
            print(f"{self.nama} sudah mati.")

    def telepon(self, nomor: str):
        if self.is_on:
            print(f"{self.nama} sedang menelepon {nomor}...")
        else:
            print(f"{self.nama} mati. Hidupkan dulu untuk menelepon.")

    def tampilkan_info(self):
        super().tampilkan_info()
        print(f"Sistem Operasi: {self.sistem_operasi}")


class Tablet(PerangkatGenggam):
    def __init__(self, nama: str, harga: float, kapasitas_baterai_mah: int, ukuran_layar_inci: float):
        super().__init__(nama, harga, kapasitas_baterai_mah)
        self.__ukuran_layar_inci = ukuran_layar_inci

    @property
    def ukuran_layar_inci(self) -> float:
        return self.__ukuran_layar_inci

    @ukuran_layar_inci.setter
    def ukuran_layar_inci(self, ukuran: float):
        if 7.0 <= ukuran <= 15.0:
            self.__ukuran_layar_inci = ukuran
        else:
            print("Ukuran layar tablet tidak valid (7.0-15.0 inci).")

    def hidupkan(self):
        if not self.is_on:
            print(f"{self.nama} sedang dihidupkan. Siap untuk produktivitas atau hiburan!")
            self._set_is_on(True)
        else:
            print(f"{self.nama} sudah menyala.")

    def matikan(self):
        if self.is_on:
            print(f"{self.nama} sedang dimatikan. Sampai jumpa!")
            self._set_is_on(False)
        else:
            print(f"{self.nama} sudah mati.")

    def buka_aplikasi(self, nama_aplikasi: str):
        if self.is_on:
            print(f"{self.nama} membuka aplikasi: {nama_aplikasi}")
        else:
            print(f"{self.nama} mati. Hidupkan dulu untuk membuka aplikasi.")

    def tampilkan_info(self):
        super().tampilkan_info()
        print(f"Ukuran Layar: {self.ukuran_layar_inci} inci")

# --- Fungsi untuk Operasi Polimorfisme (Tidak Berubah) ---
def operasikan_perangkat_elektronik(perangkat: BarangElektronik):
    print(f"\n=== Mengoperasikan {perangkat.nama} ===")
    perangkat.tampilkan_info()
    perangkat.hidupkan()
    perangkat.matikan()

# --- FUNGSI UNTUK MENGHUBUNGKAN KE JSON ---

def baca_data_elektronik_dari_json(nama_file: str) -> list[BarangElektronik]:
    """
    Membaca data barang elektronik dari file JSON dan mengubahnya menjadi objek Python.
    """
    daftar_objek_elektronik = []
    try:
        with open(nama_file, 'r', encoding='utf-8') as file:
            data_json = json.load(file) # Memuat seluruh isi JSON

        for item_data in data_json:
            tipe_perangkat = item_data.get("tipe")
            nama = item_data.get("nama")
            harga = item_data.get("harga")

            if tipe_perangkat == "Televisi":
                daya_watt = item_data.get("daya_watt")
                ukuran_layar_inci = item_data.get("ukuran_layar_inci")
                objek = Televisi(nama, harga, daya_watt, ukuran_layar_inci)
            elif tipe_perangkat == "Kulkas":
                daya_watt = item_data.get("daya_watt")
                kapasitas_liter = item_data.get("kapasitas_liter")
                objek = Kulkas(nama, harga, daya_watt, kapasitas_liter)
            elif tipe_perangkat == "Smartphone":
                kapasitas_baterai_mah = item_data.get("kapasitas_baterai_mah")
                sistem_operasi = item_data.get("sistem_operasi")
                objek = Smartphone(nama, harga, kapasitas_baterai_mah, sistem_operasi)
            elif tipe_perangkat == "Tablet":
                kapasitas_baterai_mah = item_data.get("kapasitas_baterai_mah")
                ukuran_layar_inci = item_data.get("ukuran_layar_inci")
                objek = Tablet(nama, harga, kapasitas_baterai_mah, ukuran_layar_inci)
            else:
                print(f"Peringatan: Tipe perangkat tidak dikenal: {tipe_perangkat}")
                continue
            daftar_objek_elektronik.append(objek)

    except FileNotFoundError:
        print(f"Error: File '{nama_file}' tidak ditemukan.")
    except json.JSONDecodeError:
        print(f"Error: Format JSON di '{nama_file}' tidak valid.")
    except Exception as e:
        print(f"Terjadi kesalahan saat membaca file JSON: {e}")

    return daftar_objek_elektronik

def tulis_data_elektronik_ke_json(daftar_objek: list[BarangElektronik], nama_file: str):
    """
    Menulis data objek BarangElektronik ke file JSON.
    Catatan: Metode ini harus mengekstrak data dari objek.
    """
    data_untuk_ditulis = []
    for obj in daftar_objek:
        item_data = {
            "nama": obj.nama,
            "harga": obj.harga
            # is_on tidak perlu disimpan jika defaultnya mati atau diatur saat inisialisasi
            # karena ini adalah status runtime, bukan konfigurasi dasar
        }
        if isinstance(obj, Televisi):
            item_data["tipe"] = "Televisi"
            item_data["daya_watt"] = obj.daya_watt
            item_data["ukuran_layar_inci"] = obj.ukuran_layar_inci
        elif isinstance(obj, Kulkas):
            item_data["tipe"] = "Kulkas"
            item_data["daya_watt"] = obj.daya_watt
            item_data["kapasitas_liter"] = obj.kapasitas_liter
        elif isinstance(obj, Smartphone):
            item_data["tipe"] = "Smartphone"
            item_data["kapasitas_baterai_mah"] = obj.kapasitas_baterai_mah
            item_data["sistem_operasi"] = obj.sistem_operasi
        elif isinstance(obj, Tablet):
            item_data["tipe"] = "Tablet"
            item_data["kapasitas_baterai_mah"] = obj.kapasitas_baterai_mah
            item_data["ukuran_layar_inci"] = obj.ukuran_layar_inci

        data_untuk_ditulis.append(item_data)

    try:
        with open(nama_file, 'w', encoding='utf-8') as file:
            json.dump(data_untuk_ditulis, file, indent=2, ensure_ascii=False)
        print(f"\nData berhasil disimpan ke '{nama_file}'.")
    except Exception as e:
        print(f"Terjadi kesalahan saat menulis ke file JSON: {e}")

# --- BAGIAN MAIN UNTUK DEMONSTRASI ---
if __name__ == "__main__":
    json_file_name = "daftar_elektronik.json"

    # --- Bagian 1: Membuat file JSON awal (jika belum ada) ---
    print("--- Membuat atau Memastikan File JSON Awal ---")
    data_awal = [
        {
            "tipe": "Televisi",
            "nama": "LG OLED C2",
            "harga": 15000000.0,
            "daya_watt": 150,
            "ukuran_layar_inci": 65
        },
        {
            "tipe": "Kulkas",
            "nama": "Sharp Megamouth",
            "harga": 6500000.0,
            "daya_watt": 100,
            "kapasitas_liter": 300
        },
        {
            "tipe": "Smartphone",
            "nama": "Samsung Galaxy S24 Ultra",
            "harga": 22000000.0,
            "kapasitas_baterai_mah": 5000,
            "sistem_operasi": "Android"
        },
        {
            "tipe": "Tablet",
            "nama": "iPad Pro M4",
            "harga": 18000000.0,
            "kapasitas_baterai_mah": 10000,
            "ukuran_layar_inci": 12.9
        },
        {
            "tipe": "Televisi",
            "nama": "Samsung Crystal UHD",
            "harga": 5000000.0,
            "daya_watt": 100,
            "ukuran_layar_inci": 50
        }
    ]
    try:
        with open(json_file_name, 'w', encoding='utf-8') as f:
            json.dump(data_awal, f, indent=2, ensure_ascii=False)
        print(f"File '{json_file_name}' berhasil dibuat/diperbarui dengan data awal.")
    except Exception as e:
        print(f"Gagal membuat/memperbarui file JSON: {e}")


    # --- Bagian 2: Membaca data dari JSON dan membuat objek ---
    print("\n--- Membaca Data dari JSON dan Membuat Objek Barang Elektronik ---")
    koleksi_elektronik_dari_json = baca_data_elektronik_dari_json(json_file_name)

    if koleksi_elektronik_dari_json:
        print("\nObjek yang berhasil dibuat dari JSON:")
        for item in koleksi_elektronik_dari_json:
            item.tampilkan_info()
            print("-" * 20)

        # Demonstrasi polimorfisme dengan objek dari JSON
        print("\n--- Demonstrasi Polimorfisme dengan Objek dari JSON ---")
        for item in koleksi_elektronik_dari_json:
            operasikan_perangkat_elektronik(item)
    else:
        print("Tidak ada objek yang berhasil dimuat dari JSON.")

    # --- Bagian 3: Menambahkan objek baru dan menyimpannya kembali ke JSON ---
    print("\n--- Menambahkan Objek Baru dan Menyimpan Kembali ke JSON ---")
    # Buat objek baru
    laptop_baru = Tablet("Lenovo Tab P12 Pro", 10000000.0, 8000, 12.6) # Menggunakan Tablet sebagai representasi laptop sederhana
    mesin_cuci_baru = Kulkas("LG Smart Inverter Washer", 7000000.0, 200, 12) # Menggunakan Kulkas sebagai representasi mesin cuci sederhana (hanya contoh)

    # Tambahkan objek baru ke koleksi yang sudah ada (dari JSON)
    koleksi_elektronik_dari_json.append(laptop_baru)
    koleksi_elektronik_dari_json.append(mesin_cuci_baru) # Corrected variable name

    # Simpan kembali seluruh koleksi ke file JSON
    tulis_data_elektronik_ke_json(koleksi_elektronik_dari_json, "daftar_elektronik_update.json")

    print("\n--- Membaca Ulang Data dari File JSON yang Diperbarui ---")
    koleksi_elektronik_terbaru = baca_data_elektronik_dari_json("daftar_elektronik_update.json")
    if koleksi_elektronik_terbaru:
        for item in koleksi_elektronik_terbaru:
            item.tampilkan_info()
            print("-" * 20)
```
